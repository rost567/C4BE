## Equilibrium 
Assume the quantity of a good demanded by consumers, ùëû, can be achieved with a price for that good of $ùëù_ùê∑(ùëû) = 10000 ‚àí 750 ‚ãÖ\sqrt{ùëû}$, while the supplier will charge a price of  ${ùëù_ùëÜ(ùëû) = 100 + 0.3q^2 + 0.4ùëû ^ {1.2}}$  when the quantity is ùëû and $ ùëû \ge 0$.

**a)** Implement lambda functions for $ùëù_ùê∑(ùëû)$ and $ùëù_ùëÜ(ùëû)$.<br>
**b)** Create a numpy array with quantities ranging from 0 to 200 (e.g., using np.arange or np.linspace) and visualize the two functions with the plt.plot() method.

```python
q = np.arange(201)
p_demand = list(map(lambda q: 10000 - 750 *q**(0.5), q))
p_supply = list(map(lambda q: 100+0.3*q**(2)+0.4*q**(1.2),q))

plt.plot(q, p_demand, 'b',q,p_supply, 'g')
plt.title('Price based on Demand and supply')
plt.xlabel('quantity')
plt.ylabel('price')
plt.legend(['demand', 'supply'], loc=0)


```

   >_iii) What is the equilibrium price $ùëù‚àó$ where $ùëù_ùëÜ(ùëû*) = ùëù_ùê∑(ùëû‚àó)$, and what is the corresponding quantity, $ùëû‚àó$?_

```python
equi = lambda q: 10000 - 750 *q**(0.5) - (100+0.3*q**(2)+0.4*q**(1.2)) #to find the intersection of demand and supply
q_equi = bisection_root(equi,[80,100])
p_equi = q_supply(q_equi)
print(f'bisection method:q*={q_equi},p*={p_equi}')
```

```python
interpol_q_equi = regula_falsi(equi,[80,100])
interpol_p_equi = q_supply(interpol_q_equi)
print(f'linear interpolation:q* = {interpol_q_equi},p*={interpol_p_equi}')
```

```python
newton_q_equi = newton_root(equi,110)
newton_p_equi = q_supply(newton_q_equi)
print(f'newton method: q*={newton_q_equi},p*={newton_p_equi}')
```

## Present Value 
The present value of a future payment $ùê∂_ùë°$ is $ùëÉùëâ(ùê∂_ùë°)=\frac {ùê∂_t}{(1+ùëò)^t}$ where ùë° is the
time until the payment, and ùëò is the discount (interest) rate. Unless stated
otherwise, assume ùê∂ùë° = 100, ùëò = 0.05, and ùë° = 1.

> **a)** Implementing a function PV that accepts appropriate inputs and returns the present value.

```python
def pvFunction(k=0.05, t=1, Ct=100):#default values to k,t & Ct
    PV=Ct/(1+k)**t
    return PV
print('present value=',pvFunction())#with default values
```

>**b)** Assuming $ùë° ‚àà [0, 1, 2, 3, 4, ..., 10]$. Using list comprehension, create list that contains the corresponding present values and visualize the present values with respect to ùë°.


```python
t = list(range(11))
pv_t = [pvFunction(0.05,t) for t in t] #list comprehension method
print('t=',t,'present value(t)=',pv_t)
```

> **c)** Assuming $ùëò ‚àà [0, 0.01, ..., 0.10]$. Again, create a list with the corresponding present values.

```python
k=list(map(lambda x:x/100.0, range(11)))
pv_k = [pvFunction(k) for k in k]
print('k=',k,'present value(k)=',pv_k)
```

>d) Nested lists with two variables can be created in the fashion


```python
pv_tk = [[pvFunction(k,t) for t in t] for k in k]
print('k=',k,'t=',t,'present value(k,t)=',pv_tk)
plt.contour(pv_tk)
plt.title('Contour plot for t and k')
```

## Hasse's & Collatz's algorithm  
Assume ùë•0 is an arbitrary positive integer and is the first number in a series. The next numbers are found iteratively (Hasse‚Äôs algorithm):  

$ùë•_ùë°+1 =\begin{cases} \frac{ùë•_ùë°}{2} & \text{if ùë•_ùë° is even} \\ 3 ‚ãÖ ùë•_ùë° + 1 & \text{if ùë•_ùë° is odd, i.e., 1,3,5,...} \end{cases}$ 

The Collatz conjecture states that a sequence (ie., $ùë•ùë°+1 = ùëì (ùë•ùë°))$, where this operation is performed on itself, eventually will reach 1 for any arbitrary ùë•0 that is a positive integer. Examples: 
$ùë•_0 = 4 ‚à∂ 4 \xrightarrow[x\setminus2]{\text{even}} 2 \xrightarrow[x\setminus2]{\text{even}} 1$  
  
$ùë•_0 = 3 : 3 \xrightarrow[3x+1]{\text{odd}} 10 \xrightarrow[x\setminus2]{\text{even}} 5 \xrightarrow[3x+1]{\text{odd}} 16 \xrightarrow[x\setminus2]{\text{even}} 8 \xrightarrow[x\setminus2]{\text{even}} 4 \xrightarrow[x\setminus2]{\text{even}} 2 \xrightarrow[x\setminus2]{\text{even}}1$  

$ùë•_0 = 10: 10 \xrightarrow[x\setminus2]{\text{even}} 5 \xrightarrow[3x+1]{\text{odd}} 16 \xrightarrow[x\setminus2]{\text{even}} 8 \xrightarrow[x\setminus2]{\text{even}} 4 \xrightarrow[x\setminus2]{\text{even}} 2 \xrightarrow[x\setminus2]{\text{even}}1 $ 


> a) Implementing a (regular) function Hasse(x) that accepts an integer value ùë•ùë° and returns it successor ùë•ùë°+1. (Ensure that the argument is never less or equal to zero.)

```python
def hasse(x):
    if x <= 0:#never has to be less or equal to 0
        x=False
        print('type x > 0')
    elif x % 2 == 0:
        x = x/2
    elif x % 2 == 1:
        x = 3 * x + 1
    return x
       
```

> **b)** Implementing a function Collatz$(x0)$ that accepts the first element $ùë•_0$ as input and returns a list with the resulting series as a list, and the number of steps (iterations) it took to (first) reach $1$. Example: Collatz$(4)$ should return $([4, 2, 1], 2)$, Collatz(10) should return $([10, 5, 16, 8, 4, 2, 1], 6)$, and so on.

```python
def collatz(x):
    L = [x]
    while x > 1: #it has to be grather than because if it is x>1 will be a infinite loop
        if x % 2 == 0:
            x = x/2
            L.append(x)
        elif x % 2 == 1:
            x=3 * x + 1
            L.append(x)
    return L, len(L)-1 
    
```

> **c)** For $1 \le ùë•_0 \le 10‚Ä≤000$, which creates the longest sequence before reaching 1?


```python
def reverse_collatz(x):
    L = [x]
    while x > 1:
        if x % 2 == 0:
            x = x/2
            L.append(x)
        elif x % 2 == 1:
            x = 3 * x + 1
            L.append(x)
    return len(L)-1, L
```

```python
list_collatz = [reverse_collatz(x) for x in range(1,10001)]
#the list with all the values
list_collatz.sort(reverse=True) #to see the longest sequence in the first position
print(f'the longest sequence = {list_collatz[0]}')
```
